```python
words = set()
query_words = set()

with open('wordles.txt', 'r') as f:
    file = f.readlines()
    for word in file:
        if word[0] != '#':
            words.add(word.strip())

with open('query_word.txt', 'r') as f:
    file = f.readlines()
    cnt = 0
    for word in file:
        if word[0] != '#':
            query_words.add(word.strip())

print(len(words))
print(len(query_words))
```

```python
import math

# 0: None(Gray)
# 1: Character is in answer but wrong position(Yellow)
# 2: Character is in answer and right position(Green)

# matching: compute the 0, 1, 2
def matching(answer, word):
    result = 0
    for i in range(5):
        result *= 3

        if answer[i] == word[i]:
            result += 2

        elif word[i] in answer:
            result += 1

    return result

# Node: generate the tree
class Node():
    def __init__(self, word_set):
        self.word_set = word_set
        self.word_cnt = [0] * pow(3, 5)

    def partition(self):
        max_information_gain = -1
        max_information_word = ''
        S = len(self.word_set)
        for partition_candidate in query_words:
            cnt = [0] * pow(3, 5)
            for word in self.word_set:
                cnt[matching(partition_candidate, word)] += 1
            information_gain = 0
            for i in range(pow(3, 5)):
                p = cnt[i] / S
                if p != 0:
                    information_gain -= p * math.log2(p)

            if information_gain > max_information_gain:
                max_information_gain = information_gain
                max_information_word = partition_candidate

        self.partition_word = max_information_word


    def guess(self):
        return self.partition_word

    def rt(self, resp):
        self.rt = set()
        for word in self.word_set:
            if matching(word, self.partition_word) == resp:
                self.rt.add(word)
        return self.rt

def zot_to_three(num):
    rt = 0
    for i in range(5):
        rt *= 3
        rt += int(num[i])

st = Node(words)
while True:
    st.partition()
    print(st.guess())
    resp = matching('grave', st.guess())
    if resp == pow(3, 5) - 1:
        print('Find!')
        print('Answer:', st.guess())
        break
    ns = st.rt(resp)
    if len(ns) == 1:
        print('Find!')
        print('Answer:', ns)
        break
    st = Node(ns)
```
